### MatPlotLib Library

```python
import matplotlib.pyplot as plt
```

### Linear Graph

```python
x = [1,3,5,7,11]
y = [2,4,6,8,10]

plt.plot(x,y)
plt.show()

# Syntax -> plt.plot(x-axis data, y-axis data)
# plt.show() : for remove the description
```

```python
x = [1,3,5,7,11]
y = [2,4,6,8,10]
c = 'g' 

plt.plot(x,y,color = 'r')
# plt.plot(x-axis, y-axis, color)
plt.show()
```

```python
# Random Values
x_data = [45,10,2,3,7,9]
y_data = [71,20,3,89,41,2]
plt.plot(x_data,y_data,'y',marker = 'o')
plt.show()
```

```python
# Plot the Separate graph for each axis
# Random Values
x_data = [45,10,2,3,7,9]
y_data = [71,20,3,89,41,2]
plt.plot(x_data,y_data)
plt.plot(y_data,marker = '>')
plt.plot(x_data,marker = 'o')
plt.show()
```

### Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']

plt.bar(x,y,color = c)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']

plt.barh(x,y,color = c)
plt.show()
```

```python
import numpy as np
array1 = np.array(["A","B","C","D"])
array2 = np.array([7,5,6,8])
color_list = ['r','g','b','y']

plt.barh(array1, array2, color = color_list)
plt.show()
```

```python
array1 = np.array(["A","B","C","D"])
array2 = np.array([7,5,6,8])
color_list = ['r']

plt.bar(array1, array2, color = color_list)
plt.show()
```

```python
array1 = np.array(["A","B","C","D"])
array2 = np.array([7,5,6,8])
color_list = ['g']

plt.bar(array1, array2, color = color_list, width = 0.5)
plt.show()
```

```python
array1 = np.array(["A","B","C","D"])
array2 = np.array([7,5,6,8])
color_list = ['g']
plt.title("Horizontal Bar Plot")
plt.barh(array1, array2, color = color_list, height = 0.5)
plt.show()
```

### Scatter Plot

```python
x_data = [45,10,2,3,7,9]
y_data = [71,20,3,89,41,2]
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.scatter(x_data,y_data)
plt.show()
```

```python
x_data = [45,10,2,3,7,9]
y_data = [71,20,3,89,41,2]
color_list = ['r','g','b','y','m','brown']

plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.scatter(x_data,y_data, color = color_list)
plt.show()
```

```python
x_data = [45,10,2,3,7,9]
y_data = [71,20,3,89,41,2]
color_list = ['r','g','b','y','m','brown']

plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Sales")

plt.scatter(x_data,y_data, color = color_list, s = 150 )
# Syntax = plt.scatter(x_axis, y_axis, Color=[color_list], s= (Size))
plt.show()
```
