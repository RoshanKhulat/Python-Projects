```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_number_of_Internet_users"

web = requests.get(url)
```

```python
text_conversion = BeautifulSoup(web.text,"html")
```

```python
table_list = text_conversion.find_all("table")
```

```python
len(table_list)
```

```python
working_table = table_list[5]
```

```python
working_table
```

```python
table_rows = working_table.find_all("tr")
table_rows
```

```python
table_heads = None
for row_val in table_rows[:1]:
    data_val = row_val.text.strip()
    #print(data_val.strip())
    table_heads = data_val.strip().split("\n")

table_heads
```

```python
data_frame = pd.DataFrame(columns = table_heads)
```

```python
data_frame
```

```python
for data in table_rows[1:]:
    row_data = data.find_all('td')
    data_list=[]
    for data in row_data :
        individual_data = data.text.strip()
        data_list.append(individual_data)

    length = len(data_frame)
    data_frame.loc[length] = data_list
```
