```python
hotel_file["reservation_status_date"] = pd.to_datetime(hotel_file["reservation_status_date"],format='mixed')
```

```python
hotel_file.info()
```

```python
hotel_file.describe(include = 'object')
```

```python
hotel_file.isnull().sum()
```

```python
hotel_file.drop(['company','agent'], axis = 1, inplace=True)
```

```python
hotel_file.dropna(axis = 0,inplace=True)
hotel_file.isnull().sum()
```

```python
hotel_file.describe()
```

```python
hotel_file = hotel_file[hotel_file['adr']<5000] 
```

```python
cancel_per = hotel_file['is_canceled'].value_counts(normalize = True)
cancel_per
```

```python
plt.figure(figsize = (5,4))
plt.title('Reservation Status Report')
plt.bar(['Resevred','Canceled'],hotel_file['is_canceled'].value_counts(),edgecolor = 'k', width = 0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax1 = sns.countplot(x='hotel',hue='is_canceled',data=hotel_file,palette = 'Paired')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title("Reservation status in different hotel",size=20)
plt.xlabel("hotel")
plt.ylabel("number of reservation")
plt.show()
```
