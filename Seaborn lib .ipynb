```python
import matplotlib.pyplot as plt
import seaborn as sns
```

##### matplotlib

```python
var_array_1 = [10,20,30,40]
var_array_2 = [20,30,40,50]

plt.plot(var_array_1,var_array_2, color = 'r', marker = 'o')
plt.show()
```

##### Seaborn

```python
var_array_1 = [10,20,30,40]
var_array_2 = [20,30,40,50]
sns.lineplot(x = var_array_1, y = var_array_2)
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({"var_1":var_array_1 , "var_2":var_array_2})
df
```

```python
sns.lineplot(x = 'var_1' , y = 'var_2' , data = df)
plt.show()
```
