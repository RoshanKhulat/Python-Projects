## Exploratory Data Analyst

### Gender

```python
df.columns
```

```python
ax_cont = sns.countplot(x = 'Gender',data= df)

for bar in ax_cont.containers:
    ax_cont.bar_label(bar)
```

```python
Gender_Amount_data = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(x= 'Gender',y='Amount',data = Gender_Amount_data)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age Group

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender' ,data = df)
for value in ax.containers:
    ax.bar_label(value)
```

```python
Age_group_amount = df.groupby(['Age Group'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = True)
sns.barplot( x = 'Age Group', y = 'Amount' ,data = Age_group_amount)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
df.columns
```

```python
# Top 10 State by orders

state_viz_orders = df.groupby('State' , as_index = False)['Orders'].sum().sort_values(by ='Orders' , ascending = False).head(10)
sns.set(rc = {'figure.figsize':(18,5)})
sns.barplot(data = state_viz_orders, x = 'State', y ='Orders')
```

```python
# Total amount/Sales Top 10 State
Total_sales_by_States = df.groupby('State', as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sns.set(rc = {'figure.figsize':(18,5)})
sns.barplot( x = 'State', y = 'Amount', data =Total_sales_by_States)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
df.columns
```

```python
ax = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc = {'figure.figsize' :(5,5)})
for value in ax.containers:
    ax.bar_label(value)
```

```python
amount_by_marital_status = df.groupby(['Marital_Status','Gender'],  as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot( x ='Marital_Status', y ='Amount', hue = 'Gender' , data =  amount_by_marital_status)
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set(rc = {'figure.figsize':(20,6)})
ax = sns.countplot(x = 'Occupation', data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sales_by_Occupation = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Occupation' , y = 'Amount' , data = Sales_by_Occupation)
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(25,8)})
ax = sns.countplot(x = 'Product_Category' , data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Product_cat_data = df.groupby('Product_Category', as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category', y='Amount', data = Product_cat_data)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
top_10_Product_by_sales = df.groupby(['Product_ID'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending = False).head(10)
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(x='Product_ID' , y = 'Amount' , data = top_10_Product_by_sales)
```

## Conclusion

*Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
