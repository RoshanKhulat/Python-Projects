```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_banks_in_India"
web = requests.get(url)
```

```python
web
```

```python
body = BeautifulSoup(web.text,"html")
```

```python
table_list = body.find_all("table")
```

```python
len(table_list)
```

```python
public_sector_banks = table_list[4]
```

```python
public_sector_banks
```

```python
title_rows = public_sector_banks.find_all('tr')
```

```python
title_rows
```

```python
coloumns = None
coloumns = title_rows[0].find_all("th")    
```

```python
coloumns
# coloumns holds the list inside the list 
```

```python
table_head = []
# the Fatch the Inner list use the index of list
for th in coloumns:
    key = th.text.strip()
    table_head.append(key)

table_head
```

```python
data_frame = pd.DataFrame(columns = table_head)
```

```python
data_frame
```

```python
for row in title_rows[1:]:
    row_data_list = row.find_all('td')
    row_data = [data.text.strip() for data in row_data_list]
    lenght = len(data_frame)
    data_frame.loc[lenght] = row_data
    #print(row_data)
```
